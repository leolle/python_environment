```python
import pandas as pd
from pandas import Series, DataFrame
data = DataFrame(pd.read_csv('data/test_output0.csv',encoding="gbk",na_values=['--  ']))

#result = DataFrame(columns =['stock_id'] , index = data.index)

for x in data['report'].index:
    
    data['report'][x] = DataFrame(eval(data['report'][x]))
    data['report'][x]['stock_id'] = data['name'][x]

ans = data['report']
stock_list = []
for x in data.index:
    stock_list.append(ans[x])
result = pd.concat(stock_list)
result.to_csv('data/output_0.csv',encoding = "gbk")
```
