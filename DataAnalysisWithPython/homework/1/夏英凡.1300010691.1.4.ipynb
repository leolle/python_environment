```python
%matplotlib inline 
from __future__ import division

from numpy.random import randn
import numpy as np

np.set_printoptions(precision=4, suppress=True)
```

```python
def arr(Z):
    array = np.zeros((len(Z)-2,3))
    n = 0
    for i in range(0,len(Z)-2):
        for j in range(0,3):
            array[i][j] = Z[n]
            n=n+1
        n=n-2
    return array
```

```python
Z = randn(100)
print(Z)
arr(Z)
```
