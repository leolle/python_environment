```python
data1 = pd.read_csv('data/A1.csv',encoding="gbk",na_values=['--  '])
DataFrame(data1)
data2 = pd.read_csv('data/A2.csv',encoding="gbk",na_values=['--  '])
DataFrame(data2)
data3 = pd.read_csv('data/A3.csv',encoding="gbk",na_values=['--  '])
DataFrame(data3)
data4 = pd.read_csv('data/A4.csv',encoding="gbk",na_values=['--  '])
DataFrame(data4)
data5 = pd.read_csv('data/A5.csv',encoding="gbk",na_values=['--  '])
DataFrame(data5)
```

```python
data_merge0 = pd.merge(data1, data2) 
data_merge1 = pd.merge(data_merge0,data3)
data_merge2 = pd.merge(data_merge1,data4)
data_merge  = pd.merge(data_merge2,data5)
##为下面计算加权做准备
data_merge['total1'] = data_merge[u'总资产(万)'] *data_merge[u'市盈(动)']
data_merge['total2'] = data_merge[u'总资产(万)'] *data_merge[u'今开']

##把流通市值后面的单位“亿”去掉
for k in range(2780):
    data_merge[u'流通市值'].values[k] = float(data_merge[u'流通市值'].values[k][:14])

data_merge['total3'] = data_merge[u'流通市值'] *data_merge[u'市盈(动)']
data_merge['total4'] = data_merge[u'流通市值'] *data_merge[u'今开']

data_merge.columns
```

```python
##按两种方法分类先
group1 = data_merge.groupby(u'细分行业')
group2 = data_merge.groupby(u'地区')
```
