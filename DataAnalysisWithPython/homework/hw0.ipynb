　　这次作业我们将简单讨论先天与后天，基因与环境之间的关系。
    
　　首先我们将这两个问题分开来谈一下。基因与环境在我看来是互相影响的，在生物进化的过程中，外在环境会影响基因的演变更替，比如达尔文随贝格尔号航行，在加拉帕格斯群岛上观测到的鸟类种类，其长喙短喙的数量便是鸟类基因的一种体现。而长此以往，具有某种长度的喙的鸟类一旦在岛上成为主导力量，势必会导致这个岛的整个生态环境的变化，从而导致别的生物的基因需要为此作出相应的改变。 再比如，人类在漫长的农耕时代中，驯服了家禽和家畜，这导致了这些生物基因较之于其祖先生物，或者说进化树上的前辈们，发生了一些变化，而这些变化又会影响到人类和家禽家畜共同生活的环境，比如使得更多的生物被驯服为家禽。反过来，这些环境的改变又会影响到被驯服的生物的基因表现,比如说使得狗变得更忠于人类。
   
　　所以说，不管从哪个方面来看，基因与环境就类似于一个环，但不是一个封闭的环，而是一个会受到外部冲击的开放的环。这里我们说的基因是某个生物种群的基因，环境是其所处的小范围的，排除了外生冲击来源的环境。在我们现在讨论的这样的基因与环境所处的开放环上，基因与环境都会会受到外生冲击,在外生冲击之后，一段时间内基因与环境都会对此外生冲击作出反应，并且该反应会体现出外生冲击的影响。我们可以用一个AR(1)时间序列的模型来说明这一点。


$$Y_{t} = \alpha_{1} X_{t-1} + \alpha_{2}  hit_{t-1} + \epsilon_{t}$$



$$X_{t} = \beta_{1} Y_{t-1} + \beta_{2} hit_{t-1} + e_{t}$$



　　其中，$Y_{t}$表示了环境在 $t$ 时期的表现，$X_{t}$表示了基因 在 $t$ 时期的表现，$hit_{t-1}$ 即为我们在某一时期受到的外生冲击，当然可以简单地认为这是一个$i.i.d$的序列。公式最后的项即为白噪声项。
 
　　毫无疑问地，这个AR（1）模型是不够严谨的，首先AR（1）的假设就过于武断，环境与基因的影响在现实中是绝不可能仅仅局限于一期的。另外简单的线性模型也是不够严谨的，生物的基因及其表现也不会很快地在一期时间内就发生可观测到的变化，而且我还在没有实际的参数拟合之前作出了函数形式的假设。但我们将模型列在这里，并不是为了扯淡，而是可以看到这两者之间的互相影响的关系。并且这两者也是会受到外生冲击$hit_{t}$的，这体现了不管生物体在怎样的一个生态环境之中，小环境和基因都是躲不开更大环境的影响。即使大如整个地球的生物圈，一旦受到外太空的影响，如太阳风暴，或者小行星冲击，也是会在外生冲击的刺激下发生相关的变化。

　　具体到人而言，人从灵长类动物之中进化而来，在适应地球生活环境的同时，自己的基因结构发生了变化，作为基因的外现，人类有了更发达的大脑，作为代价，人类牺牲了肌肉力量，毛发，利齿等。而反过来，人类在进化的过程中，也在不断改造自然，影响着生存的环境。地表的植被覆盖率从工业革命以来已经减少了40%以上，而空气的主要成分如氮气、氧气的比例相较于地球的原始大气，也早就发生了巨大的变化，而且可以预见到二氧化碳的浓度也会在未来继续上升。环境--生物圈，和基因--人类，就是在这样一个循环之中互相影响。倘若有一次巨大的外生冲击，如某小行星在未来几十年内击中地球，我们又可以将其列入外生冲击项来继续我们的讨论了（人类还没因此灭绝的话）。

　　所以说，这个模型虽然简陋，但是却可以深刻地反应出环境、生物基因的互相影响。
  
　　然后我们再讨论一下先天与后天的关系。
　　
　　先天是什么？在我看来，毫无疑问的，生物的先天条件部分决定了其性状特征以及生活习惯。如狼吃肉，羊吃草，这些都是先天基因决定的，在生物自己的主观范畴之外的因素先决地确定了这一些特征。这些不可抗拒力量决定的事情会使得生物在齐生命周期之内之内沿着基因谱写好的剧本进行生活。倘若一只羊想吃肉，或是说一只狼热衷于食草，那么他们的生命将注定短暂，违抗了生物规律的生存都是短寿的。

　　而对应的，后天则是生物自身在环境中受到的影响，以及生物自身对这样的影响作出的反应。我们上面刚刚提到过，生物会对环境作出反应，并且环境会将生物的反应反馈给生物本身，如此循环往复。所以我们在谈到后天时，一定要注意到，后天不仅仅是生物的主观决定，还有很多事环境的力量。

　　一个简单的例子，一位同学A出生在上海，一位同学B出生在重庆，首先他们来自不同的家庭，有不同的父母，那么他们的基因是不相同的（准确来讲，99%以上是相同的，但正是不同的部分使得两位同学成了两个完全不一样的人），这样不同的遗传信息就是不一样的先天条件。或许B同学继承了父亲爱吃辣的口味，A同学从母亲那里继承了爱吃甜的口味。这样不一样的先天条件在两位同学出生之前就决定了他们生活的不同。而在他们出生之后，A同学在走街串巷的同时学会了一口流利的上海话，B同学则耳濡目染，练就了出色的麻将技巧。沿海大城市上海让A同学看到了世博会的召开，参加了学校对外交流的很多项目，认识了不少外国同学。山城崎岖的地形让B同学从小就对从头顶飞驰而过的车辆见怪不怪，学习之余的空闲时间回和同学一起去泡个澡喝个茶。由此可见，后天的环境让两位同学的人生轨迹又有了更大的差别，两位同学的生活飞驰在两条不同的道路上。

　　而先天与后天的关系，在我看来也是互相影响，互相决定的。一些具有相类似的先天条件的个体聚集在一起，会形成一种环境，而这种环境则会影响到生活在这个环境中的个体的后天条件，甚至也会影响到诞生在这个环境中的个体的先天条件。用一个大家都能理解的例子来讲，每年全国最优秀最聪明的孩子们会聚集到清华大学和北京大学，或者更广泛些，聚集到北京海淀这样一个科教中心。这就形成了一个环境，在这个环境中生活的人会有一些共同点，比如尊重知识，科学理性。也会受到一些冲击的共同影响，比如高校不提前供暖会使得同学们在十月底冷得瑟瑟发抖。然后这样的一个环境内会有不同的男男女女认识，交往，结婚生子。这样他们的孩子又会继承父母的优秀基因，获得了其他地方的人很难获得的先天条件。
  
　　不难发现，生物的先天与后天的关系，和基因与环境的关系是非常类似的，我们其实也可以用一个时间序列的模型来描述这样的关系，只不过为了简单起见，加之之前的模型过于粗浅，作者就不再次提起。

　　这样的类似背后是有生物遗传的深层次原因的，其实我们可以跳出个体来看这个问题。心理学上讲，生物个体都是基因的载体，我们与我们的父辈祖辈享有共同的基因，从基因层面来看，我们和他们是一样的，只是基因的载体不一样。而基因就是生物的先天条件，环境就是生物生存的后天条件。所以生物的基因与环境的关系，是完全类比于先天与后天的关系的。跳出生物个体的限制，从更为宏观的角度来看，两者完全类似。这就不能解释两对关系的相似性了。
